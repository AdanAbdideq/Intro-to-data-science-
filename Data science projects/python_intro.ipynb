#**OOP keywords**
- **Self keyword**
       when you want to intialize two or more objects of the same class, python needs a way to differenciate between them
       self - pointer to the memory address where each object of the class is stored
  
- **Class keyword**
      class keyword is used to define a class
      its conventional to use ClassName for the class

    

- **init method**
      - this method stores the intial values for all the instance attribute
      - we create init function like this __init__()
      - the first argument must be slef
      - we access the attribute using dot


example of a class

```python
class Car:
  def __init__(self, color, year, model):
    self.color = color
    self.year = year
    self.model = model
```

### **class vs instance attribute**
      The attribute inside init method are specific fro the instance
      called instance attribute
      Attribute that are shared between all the instance are called class attribute and are defined outside the __init__ method

      e.g
      class Car:
        def __init__(self, color, year, model):
          self.color = color
          self.year = year
          self.model = model
        number_wheels = 4
        
    number of wheels in this case are class attribute since they are defined outside the class __init__ method

```python
class Car:
  def __init__(self, year):
    self.year = year
  number_wheels = 4
```

```python
car_2021 = Car(2021)
print(car_2021.year)  # instance attribute
print(car_2021.number_wheels)   # class attribute
```
